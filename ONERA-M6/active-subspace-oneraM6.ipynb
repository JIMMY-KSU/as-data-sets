# Active Subspace Analysis Notebook for ONERA M6 Shape Optimization

**DESCRIPTION**

This data set comes from a simulation of the ONERA M6 transonic wing design. The inputs are parameters of the free-form deformation boxes for the wing geometry. The outputs are the lift and drag of the wing. Gradients are available from an adjoint solver. All simulations are performed with [SU2](http://su2.stanford.edu/).

**INPUTS (50)**

Variable | Lower bound | Upper bound | Density
--- | --- | --- | ---
x1-x50 | -0.05 | 0.05 | Uniform

**OUTPUTS (2)**

Variable | Description
--- | --- 
Lift | The lift of the wing design.
Drag | The drag of the wing design.

**REFERENCES**

+ Lukaczyk, Palacios, Alonso, and Constantine. [Active Subspaces for Shape Optimization](http://arc.aiaa.org/doi/abs/10.2514/6.2014-1171)
+ Constantine, Dow, and Wang. [Active Subspace Methods in Theory and Practice: Applications to Kriging Surfaces](http://epubs.siam.org/doi/abs/10.1137/130916138)

```python
import numpy as np
import pandas as pn
import active_subspaces as ac
```

Import the data set. Distinguish inputs (X), outputs (F), and gradients (G).

```python
df = pn.DataFrame.from_csv('ONERA-M6.txt')
data = df.as_matrix()
X = data[:,:50]
F = data[:,50:52]
G_lift = data[:,52:102]
G_drag = data[:,102:]
M,m = X.shape
labels = df.keys()
out_labels = labels[50:52]
```

Normalize the inputs to the interval [-1,1].

```python
xl = -0.05*np.ones((1,m))
xu = 0.05*np.ones((1,m))
XX = ac.normalize_uniform(X,xl,xu)
```

Choose the output of interest.

```python
out_index = 1 # either 0 (lift) or 1 (drag)
f = F[:,out_index]
if out_index == 0:
    G = G_lift
else:
    G = G_drag
out_label = out_labels[out_index]
```

Compute the active subspace.

```python
k = 6 # number of eigenvalues 
lam,W,lam_bootrange,sub_bootrange = ac.get_active_subspace(G,k)
```

Plot eigenvalues and subspace errors.

```python
ac.plot_active_subspace(lam,W,
                        lam_bootrange=lam_bootrange,
                        sub_bootrange=sub_bootrange,
                        out_label=out_label)
```

One- and two-d sufficient summary plots.

```python
y = np.dot(XX,W[:,:2])
w = W[:,:2]
ac.sufficient_summary_plot(y,f,w,out_label=out_label)
```
